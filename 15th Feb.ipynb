# Q1. What is multiprocessing in python? Why is it useful? 

### Multiprocessing refers to the ability of a system to support more than one processor at the same time. Applications in a multiprocessing system are broken to smaller routines that run independently each with its own memory space and Python interpreter. The operating system allocates these threads to the processors improving performance of the system.

### advantages of multiprocessing are :
### Multiprocessing in Python is an effective mechanism for memory management. It enables you to create programs that bypass the GIL and make optimum use of your CPU core.
### Multiprocessing eliminates the issues associated with threading such as deadlocks and data corruption. 
### Each process has its own memory space, reducing potential memory conflicts.

# Q2. What are the differences between multiprocessing and multithreading?

### 1- Python multithreading is a technique where multiple threads are generated by a single process at the same time.
###    Python multiprocessing is a technique where multiple processes run across multiple cores at the same time.

### 2- In multithreading The GIL does not allow the threads to run simultaneously.
###    In multiprocessing Each program has its own interpreter that executes different processes at the same time.

### 3- Multithreading gives a feeling that threads are running simultaneously, but they work concurrently.
###    In multiprocessing Multiple processes actually run parallelly as the multiprocessing module streamlines the
###    independent processes by using sub-processes.

### 4- Python multithreading implements concurrency.
###    Python multiprocessing implements parallelism.

# Q3 Write a python code to create a process using the multiprocessing module.

```python
import multiprocessing
def test():
    print("this is my multiprocessing program")

if __name__ == "__main__":
    # create a child program into main program
    m= multiprocessing.Process(target=test)
    print("this is my main program")
    #starts child process
    m.start()
    #wait untill= child process terminates
    m.join()
```

# Q4 What is a multiprocessing pool in python? Why is it used?

### Python multiprocessing Pool can be used for parallel execution of a function across multiple input values, distributing the input data across processes (data parallelism).The pool is used because it can provide a generic interface for executing ad hoc tasks with a variable number of arguments, much like the target property on the Process object, but does not require that we choose a process to run the task, start the process, or wait for the task to complete.

# Q5. How can we create a pool of worker processes in python using the multiprocessing module?

```python
# code below is to create a pool of worker processes numbering 40 
import multiprocessing

from multiprocessing.pool import Pool
 
# protect the entry point
if __name__ == '__main__':
    # create a process pool with many workers
    with Pool(40) as pool:
        # report the status of the pool
        print(pool)
```

# Q6. Write a python program to create 4 processes, each process should print a different number using the multiprocessing module in python.
